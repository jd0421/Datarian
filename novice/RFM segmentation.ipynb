```python
# zip 파일 unzip

import zipfile

# 압축 파일 경로
zip_file_path = "us-ecommerce-record-2020.zip"
extract_to_path = "./unzipped_folder"

# 압축 해제
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_path)

print("압축 해제 완료!")
```
